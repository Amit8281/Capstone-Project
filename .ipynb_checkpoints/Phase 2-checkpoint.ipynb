# SQL

### Make the table in  csv format  (3 csv files)

**Data1.csv :**   sno, Hotel name ,cities, location, number of reviews, user ratings, star rating, review score

**Data2.csv  :** sno, Hotel name ,Free wifi, Family rooms ,Non Smoking Rooms, Restaurant ,Bar, Heating ,Lift ,Breakfast Cuisine1,Breakfast Cuisine2, Breakfast Cuisine3

(can take Yes /NO values)

**Data3.csv :**    sno ,  Hotel name, Room Type ,Single Bed ,Double bed  ,Prices



````python
# Create Data1.csv
data1_df = df[['sno', 'Hotel_Name', 'city', 'location', 'number_of_reviews', 'user_ratings', 'Star_Rating', 'review_score']]
data1_df.to_csv('Data1.csv', index=False)

# Create Data2.csv
data2_df = df[['sno', 'Hotel_Name', 'wifi_box', 'Family_rooms', 'Non_Smoking_Rooms', 'Restaurant', 'bar', 'Heating', 'Lifts', 'Breakfast Cuisine1', 'Breakfast Cuisine2', 'Breakfast Cuisine3']]
data2_df.to_csv('Data2.csv', index=False)

# Create Data3.csv
data3_df = df[['sno', 'Hotel_Name', 'Room_type', 'Single_Bed', 'Double_bed', 'price']]
data3_df.to_csv('Data3.csv', index=False)
````

### Creating Database Named ```Hotel_Booking```

````python
# Create a SQLite database (if it doesn't exist)
conn = sqlite3.connect('Hotel_Booking.db')
cursor = conn.cursor()
````

### Creating 3 tables inside DataBase ```Hotel_Booking```  Named ```data1``` , ```data2``` and ```data3```

````python
# Create data1 table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data1 (
        sno INTEGER PRIMARY KEY,
        Hotel_Name TEXT,
        city TEXT,
        location TEXT,
        number_of_reviews INTEGER,
        user_ratings REAL,
        Star_Rating REAL,
        review_score REAL
    )
''')

# Create data2 table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data2 (
        sno INTEGER PRIMARY KEY,
        Hotel_Name TEXT,
        wifi_box TEXT,
        Family_rooms TEXT,
        Non_Smoking_Rooms TEXT,
        Restaurant TEXT,
        bar TEXT,
        Heating TEXT,
        Lifts TEXT,
        Breakfast_Cuisine1 TEXT,
        Breakfast_Cuisine2 TEXT,
        Breakfast_Cuisine3 TEXT
    )
''')

# Create data3 table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data3 (
        sno INTEGER PRIMARY KEY,
        Hotel_Name TEXT,
        Room_Type TEXT,
        Single_Bed INTEGER,
        Double_Bed INTEGER,
        Prices INTEGER
    )
''')

# Commit the changes and close the connection
conn.commit()


print("SQLite database 'Hotel_Booking' and tables 'data1,' 'data2,' and 'data3' created successfully.")
````

### Uploading all the the 3 ```csv's``` inside their respected tables

````python
# Specify the paths to your CSV files
data1_csv = 'Data1.csv'
data2_csv = 'Data2.csv'
data3_csv = 'Data3.csv'

# Read CSV files into DataFrames
data1_df = pd.read_csv(data1_csv)
data2_df = pd.read_csv(data2_csv)
data3_df = pd.read_csv(data3_csv)

# Upload data to respective tables
data1_df.to_sql('data1', conn, if_exists='replace', index=False)
data2_df.to_sql('data2', conn, if_exists='replace', index=False)
data3_df.to_sql('data3', conn, if_exists='replace', index=False)

# Commit changes and close the connection
conn.commit()

print("Data from CSV files uploaded to their respective tables in the 'Hotel_Booking' database.")
````

### 3.Write a query to count the occurrences of each breakfast cuisine type.

````python
# Define your SQL query
query = """
SELECT "Breakfast Cuisine1", COUNT(*) AS Count
FROM data2
GROUP BY "Breakfast Cuisine1"
UNION
SELECT "Breakfast Cuisine2", COUNT(*) AS Count
FROM data2
GROUP BY "Breakfast Cuisine2"
UNION
SELECT "Breakfast Cuisine3", COUNT(*) AS Count
FROM data2
GROUP BY "Breakfast Cuisine3";
"""

# Execute the query and fetch the result into a DataFrame
df = pd.read_sql_query(query, conn)



# Print the DataFrame
df
````

## From here we will be making SUBQUERIES including the data

### 1.Write a query to find details of hotels with user ratings above 5 as a  threshold and that offer free WiFi. (Use table 1 and table 2)



````python
# Define your SQL query
query = """
SELECT d1.city, d2.Hotel_Name, d1.user_ratings, d2.wifi_box
FROM data2 AS d2
JOIN data1 AS d1 ON d2.Hotel_Name = d1.Hotel_Name
WHERE d1.user_ratings > 5
  AND d2.wifi_box = 'yes';
"""

# Execute the query and fetch the result into a DataFrame
df = pd.read_sql_query(query, conn)



# Print the DataFrame
df


````

### 2. Write a query to find hotels that offer both family rooms and non-smoking rooms(Use table 1 and Table 2)

````python
# Define your SQL query
query = """
SELECT  d1.Hotel_Name, d2.Family_rooms, d2.Non_Smoking_Rooms
FROM data1 AS d1
JOIN data2 AS d2 ON d1.Hotel_Name = d2.Hotel_Name
WHERE d2.Family_rooms = 'yes'
  AND d2.Non_Smoking_Rooms = 'yes';
"""

# Execute the query and fetch the result into a DataFrame
df = pd.read_sql_query(query, conn)



# Print the DataFrame
df
````

###  3.Write a query to calculate the average price for each room type (Use table2 and Table 3)

````python
# Define your SQL query
query = """
SELECT d3.Room_Type, AVG(d3.price) AS Average_Price
FROM data3 AS d3
JOIN data2 AS d2 ON d3.Hotel_Name = d2.Hotel_Name
GROUP BY d3.Room_Type;
"""

# Execute the query and fetch the result into a DataFrame
df = pd.read_sql_query(query, conn)


# Print the DataFrame
df
````
