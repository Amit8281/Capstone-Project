```python
Hotel_Name=[]
wifi_box = [] #done
location = [] #done
number_of_reviews = []#done
user_ratings = [] #done
star_rating = [] #done
review_score = [] # done
Family_rooms = [] #done
Non_Smoking_Rooms = [] # done 
Restaurant = [] # done 
bar = [] #done 
Heating = [] # done 
Lifts = [] # done
Breakfast = [] # done
#Room_Type = []
Single_Bed = [] # done 
Double_bed = [] # done
#Prices = []

for link in hotel_links[:500]:
    product_request = requests.get(link)
    product_html = bs(product_request.text, 'html.parser')
    
    ######################################## Hotel Name ##################################################
    name = product_html.find_all("h2", {"class": "d2fee87262 pp-header__title"})
    if name:
        hname = name[0].text.strip()
    else:
        hname = 'NA'
    # Append user_rating to the user_ratings list
    Hotel_Name.append(hname)
    
    
    ######################################## 1. wifi ########################################################
    wi = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    wifi_found = False

    for element in wi:
        text = element.get_text()
        if "Free WiFi" in text:
            wifi_found = True
            break  # Exit the loop early since we found "Free WiFi" once

    if wifi_found:
        wifi = "yes"
    else:
        wifi = "no"

    wifi_box.append(wifi)

    
    ######################################### 2. location ####################################################
    location_elements = product_html.find_all("p", {"class": "address address_clean"})
    #location_texts = []
    
    # Iterate through location elements and extract text from each
    for location_element in location_elements:
        if location_element:
            location_text = location_element.text.strip().split('\n')[0]  # Get text up to the first '\n'
        else:
            location_text = 'NA'  # Append "NA" if there is no data
        location.append(location_text)
        
    ####################################### 3. number of reviews ################################################    
    # reviews_elements should be a BeautifulSoup ResultSet
    reviews_elements = product_html.find_all("span", {"class": "a3b8729ab1 f45d8e4c32 d935416c47"})

    if reviews_elements:
        extracted_text = reviews_elements[0].text.split("·")[1].strip()
    else:
        extracted_text = 'NA'
    
    number_of_reviews.append(extracted_text)
    
    ######################################### 4. User Rating ##################################################
    
    user_rating_elements = product_html.find_all("div", {"class": "a3b8729ab1 d86cee9b25"})
    if user_rating_elements:
        user_rating = user_rating_elements[0].text.strip()
    else:
        user_rating = 'NA'
    # Append user_rating to the user_ratings list
    user_ratings.append(user_rating)
    
    ######################################### 5 . Review Score ###################################################
    score = product_html.find_all("div", {"class":"a3b8729ab1 e6208ee469 cb2cbb3ccb"})
    if score:
        user_score = score[0].text.strip()
    else:
        user_score = 'NA'
    # Append user_rating to the user_ratings list
    review_score.append(user_score)
    
    ##################################### non-smoking #######################################################
    no_smoking = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    found_non_smoking_rooms = False

    for element in no_smoking:
        text = element.get_text()
        if "Non-smoking rooms" in text:
            found_non_smoking_rooms = True
            break  # Exit the loop early since we found "Non-smoking rooms" once

    if found_non_smoking_rooms:
        non_smoking = "yes"
    else:
        non_smoking = "no"

    Non_Smoking_Rooms.append(non_smoking)

    
    
    ################################### Star Rating ########################################
    quality = product_html.find_all("div", {"class": "a53cbfa6de c70c875b96"})
    quality_score = 'no'

    if quality:
        text = quality[0].text.strip()
        match = re.search(r'(\d out of 5)', text)
        if match:
            quality_score = match.group()
    star_rating.append(quality_score)
    
    
    #################################################### Family rooms #############################################
    fam = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    found_family_rooms = False

    for element in fam:
        text = element.get_text()
        if "Family rooms" in text:
            found_family_rooms = True
            break  # Exit the loop early since we found "Family rooms" once

    if found_family_rooms:
        fam_rooms = "yes"
    else:
        fam_rooms = "no"

    Family_rooms.append(fam_rooms)

    
    
    # ###################################### restorents ##############################################
    res = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    restaurant_found = False

    for element in res:
        text = element.get_text()
        if "restaurants" in text:
            restaurant_found = True
            break  # Exit the loop early since we found "restaurants" once

    if restaurant_found:
        restaurants = "yes"
    else:
        restaurants = "no"

    Restaurant.append(restaurants)

    
    ################################################ bar #################################################
    ba=product_html.find_all("div", {"class":"hotelchars"})[0].text
    # Use a regular expression to extract "Non-smoking rooms"
    matchsss = re.search(r'Bar', ba)

    if matchsss:
        bars = matchsss.group()
    else:
        bars = 'NA'
        
    bar.append(bars)
    
    ########################################## Heating ###################################
    he = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    heat_found = False

    for element in he:
        text = element.get_text()
        if "Heating" in text:
            heat_found = True
            break  # Exit the loop early since we found "Heating" once

    if heat_found:
        heating = "yes"
    else:
        heating = "no"

    Heating.append(heating)

    
    ###################################################### Lift ############################################
    li = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    lift_found = False

    for element in li:
        text = element.get_text()
        if "Lift" in text:
            lift_found = True
            break  # Exit the loop early since we found "Lift" once

    if lift_found:
        lifts = "yes"
    else:
        lifts = "no"

    Lifts.append(lifts)
    
     ###################################### Single Bed ##################################################
    sb=product_html.find_all("div", {"class":"hotelchars"})[0].text
    # Use a regular expression to extract "Non-smoking rooms"
    matchsssss = re.search(r'single bed', sb)

    if matchsssss:
        single_bed = matchsssss.group()
    else:
        single_bed = 'NA'
        
    Single_Bed.append(single_bed)
    
    
    ############################################### double Bed ####################################################
    db=product_html.find_all("div", {"class":"hotelchars"})[0].text
    # Use a regular expression to extract "Non-smoking rooms"
    matchssssss = re.search(r'double bed', db)

    if matchssssss:
        double_bed = matchssssss.group()
    else:
        double_bed = 'NA'
        
    Double_bed.append(double_bed)
    
    ################################################### Break Fast ####################################
    
    # Find all elements that match the specified criteria
    elements = product_html.find_all("div", {"class": "bui-grid__column bui-grid__column-4 k2-hp--highlights"})

    # Check if any elements were found
    if elements:
        # If elements were found, access the first element and extract the text
        a = elements[0].text

        # Check if "Breakfast info" is present in the text
        if "Breakfast info" in a:
            # Split the text by the "Breakfast info" section and extract the breakfast information
            breakfast_section = a.split("Breakfast info")[1].strip()
            breakfast_info = breakfast_section.split("\n")[0].strip()
        else:
            breakfast_info = 'NA'
    else:
        # If no elements were found, set breakfast_info to 'NA'
        breakfast_info = 'NA'

    # Print or use the breakfast information as needed
    Breakfast.append(breakfast_info)

     
```

```python

# Create a dictionary using your lists
data = {
    'Hotel_Name':Hotel_Name,
    'wifi_box': wifi_box,
    'location': location,
    'number_of_reviews': number_of_reviews,
    'user_ratings': user_ratings,
    'Star_Rating': Star_Rating[:500],
    'review_score': review_score,
    'Family_rooms': Family_rooms,
    'Non_Smoking_Rooms': Non_Smoking_Rooms,
    'Restaurant': Restaurant,
    'bar': bar,
    'Heating': Heating,
    'Lifts': Lifts,
    'Breakfast': Breakfast,
    'Single_Bed': Single_Bed,
    'Double_bed': Double_bed,
    'Room_type':Room_type[:500],
    'price': price[:500]
}

# Create a DataFrame from the dictionary
df1 = pd.DataFrame(data)

# Optionally, you can save the DataFrame to a CSV file
df1.to_csv('London_dataset1.csv', index=False)

# Display the DataFrame
df1
```

```python
Hotel_Name=[]
wifi_box = [] #done
location = [] #done
number_of_reviews = []#done
user_ratings = [] #done
star_rating = [] #done
review_score = [] # done
Family_rooms = [] #done
Non_Smoking_Rooms = [] # done 
Restaurant = [] # done 
bar = [] #done 
Heating = [] # done 
Lifts = [] # done
Breakfast = [] # done
#Room_Type = []
Single_Bed = [] # done 
Double_bed = [] # done
#Prices = []

for link in hotel_links[:250]:
    product_request = requests.get(link)
    product_html = bs(product_request.text, 'html.parser')
    
    ######################################## Hotel Name ##################################################
    name = product_html.find_all("h2", {"class": "d2fee87262 pp-header__title"})
    if name:
        hname = name[0].text.strip()
    else:
        hname = 'NA'
    # Append user_rating to the user_ratings list
    Hotel_Name.append(hname)
    
    
    ######################################## 1. wifi ########################################################
    wi = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    wifi_found = False

    for element in wi:
        text = element.get_text()
        if "Free WiFi" in text:
            wifi_found = True
            break  # Exit the loop early since we found "Free WiFi" once

    if wifi_found:
        wifi = "yes"
    else:
        wifi = "no"

    wifi_box.append(wifi)

    
    ######################################### 2. location ####################################################
    location_elements = product_html.find_all("p", {"class": "address address_clean"})
    #location_texts = []
    
    # Iterate through location elements and extract text from each
    for location_element in location_elements:
        if location_element:
            location_text = location_element.text.strip().split('\n')[0]  # Get text up to the first '\n'
        else:
            location_text = 'NA'  # Append "NA" if there is no data
        location.append(location_text)
        
    ####################################### 3. number of reviews ################################################    
    # reviews_elements should be a BeautifulSoup ResultSet
    reviews_elements = product_html.find_all("span", {"class": "a3b8729ab1 f45d8e4c32 d935416c47"})

    if reviews_elements:
        extracted_text = reviews_elements[0].text.split("·")[1].strip()
    else:
        extracted_text = 'NA'
    
    number_of_reviews.append(extracted_text)
    
    ######################################### 4. User Rating ##################################################
    
    user_rating_elements = product_html.find_all("div", {"class": "a3b8729ab1 d86cee9b25"})
    if user_rating_elements:
        user_rating = user_rating_elements[0].text.strip()
    else:
        user_rating = 'NA'
    # Append user_rating to the user_ratings list
    user_ratings.append(user_rating)
    
    ######################################### 5 . Review Score ###################################################
    score = product_html.find_all("div", {"class":"a3b8729ab1 e6208ee469 cb2cbb3ccb"})
    if score:
        user_score = score[0].text.strip()
    else:
        user_score = 'NA'
    # Append user_rating to the user_ratings list
    review_score.append(user_score)
    
    ##################################### non-smoking #######################################################
    no_smoking = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    found_non_smoking_rooms = False

    for element in no_smoking:
        text = element.get_text()
        if "Non-smoking rooms" in text:
            found_non_smoking_rooms = True
            break  # Exit the loop early since we found "Non-smoking rooms" once

    if found_non_smoking_rooms:
        non_smoking = "yes"
    else:
        non_smoking = "no"

    Non_Smoking_Rooms.append(non_smoking)

    
    
    ################################### Star Rating ########################################
    quality = product_html.find_all("div", {"class": "a53cbfa6de c70c875b96"})
    quality_score = 'no'

    if quality:
        text = quality[0].text.strip()
        match = re.search(r'(\d out of 5)', text)
        if match:
            quality_score = match.group()
    star_rating.append(quality_score)
    
    
    #################################################### Family rooms #############################################
    fam = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    found_family_rooms = False

    for element in fam:
        text = element.get_text()
        if "Family rooms" in text:
            found_family_rooms = True
            break  # Exit the loop early since we found "Family rooms" once

    if found_family_rooms:
        fam_rooms = "yes"
    else:
        fam_rooms = "no"

    Family_rooms.append(fam_rooms)

    
    
    # ###################################### restorents ##############################################
    res = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    restaurant_found = False

    for element in res:
        text = element.get_text()
        if "restaurants" in text:
            restaurant_found = True
            break  # Exit the loop early since we found "restaurants" once

    if restaurant_found:
        restaurants = "yes"
    else:
        restaurants = "no"

    Restaurant.append(restaurants)

    
    ################################################ bar #################################################
    ba=product_html.find_all("div", {"class":"hotelchars"})[0].text
    # Use a regular expression to extract "Non-smoking rooms"
    matchsss = re.search(r'Bar', ba)

    if matchsss:
        bars = matchsss.group()
    else:
        bars = 'NA'
        
    bar.append(bars)
    
    ########################################## Heating ###################################
    he = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    heat_found = False

    for element in he:
        text = element.get_text()
        if "Heating" in text:
            heat_found = True
            break  # Exit the loop early since we found "Heating" once

    if heat_found:
        heating = "yes"
    else:
        heating = "no"

    Heating.append(heating)

    
    ###################################################### Lift ############################################
    li = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    lift_found = False

    for element in li:
        text = element.get_text()
        if "Lift" in text:
            lift_found = True
            break  # Exit the loop early since we found "Lift" once

    if lift_found:
        lifts = "yes"
    else:
        lifts = "no"

    Lifts.append(lifts)
    
     ###################################### Single Bed ##################################################
    sb=product_html.find_all("div", {"class":"hotelchars"})[0].text
    # Use a regular expression to extract "Non-smoking rooms"
    matchsssss = re.search(r'single bed', sb)

    if matchsssss:
        single_bed = matchsssss.group()
    else:
        single_bed = 'NA'
        
    Single_Bed.append(single_bed)
    
    
    ############################################### double Bed ####################################################
    db=product_html.find_all("div", {"class":"hotelchars"})[0].text
    # Use a regular expression to extract "Non-smoking rooms"
    matchssssss = re.search(r'double bed', db)

    if matchssssss:
        double_bed = matchssssss.group()
    else:
        double_bed = 'NA'
        
    Double_bed.append(double_bed)
    
    ################################################### Break Fast ####################################
    
    # Find all elements that match the specified criteria
    elements = product_html.find_all("div", {"class": "bui-grid__column bui-grid__column-4 k2-hp--highlights"})

    # Check if any elements were found
    if elements:
        # If elements were found, access the first element and extract the text
        a = elements[0].text

        # Check if "Breakfast info" is present in the text
        if "Breakfast info" in a:
            # Split the text by the "Breakfast info" section and extract the breakfast information
            breakfast_section = a.split("Breakfast info")[1].strip()
            breakfast_info = breakfast_section.split("\n")[0].strip()
        else:
            breakfast_info = 'NA'
    else:
        # If no elements were found, set breakfast_info to 'NA'
        breakfast_info = 'NA'

    # Print or use the breakfast information as needed
    Breakfast.append(breakfast_info)

     
```

```python
Hotel_Name=[]
wifi_box = [] #done
location = [] #done
number_of_reviews = []#done
user_ratings = [] #done
star_rating = [] #done
review_score = [] # done
Family_rooms = [] #done
Non_Smoking_Rooms = [] # done 
Restaurant = [] # done 
bar = [] #done 
Heating = [] # done 
Lifts = [] # done
Breakfast = [] # done
#Room_Type = []
Single_Bed = [] # done 
Double_bed = [] # done
#Prices = []

for link in hotel_links[:800]:
    product_request = requests.get(link)
    product_html = bs(product_request.text, 'html.parser')
    
    ######################################## Hotel Name ##################################################
    name = product_html.find_all("h2", {"class": "d2fee87262 pp-header__title"})
    if name:
        hname = name[0].text.strip()
    else:
        hname = 'NA'
    # Append user_rating to the user_ratings list
    Hotel_Name.append(hname)
    
    
    ######################################## 1. wifi ########################################################
    wi = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    wifi_found = False

    for element in wi:
        text = element.get_text()
        if "Free WiFi" in text:
            wifi_found = True
            break  # Exit the loop early since we found "Free WiFi" once

    if wifi_found:
        wifi = "yes"
    else:
        wifi = "no"

    wifi_box.append(wifi)

    
    ######################################### 2. location ####################################################
    location_elements = product_html.find_all("p", {"class": "address address_clean"})
    #location_texts = []
    
    # Iterate through location elements and extract text from each
    for location_element in location_elements:
        if location_element:
            location_text = location_element.text.strip().split('\n')[0]  # Get text up to the first '\n'
        else:
            location_text = 'NA'  # Append "NA" if there is no data
        location.append(location_text)
        
    ####################################### 3. number of reviews ################################################    
    # reviews_elements should be a BeautifulSoup ResultSet
    reviews_elements = product_html.find_all("span", {"class": "a3b8729ab1 f45d8e4c32 d935416c47"})

    if reviews_elements:
        extracted_text = reviews_elements[0].text.split("·")[1].strip()
    else:
        extracted_text = 'NA'
    
    number_of_reviews.append(extracted_text)
    
    ######################################### 4. User Rating ##################################################
    
    user_rating_elements = product_html.find_all("div", {"class": "a3b8729ab1 d86cee9b25"})
    if user_rating_elements:
        user_rating = user_rating_elements[0].text.strip()
    else:
        user_rating = 'NA'
    # Append user_rating to the user_ratings list
    user_ratings.append(user_rating)
    
    ######################################### 5 . Review Score ###################################################
    score = product_html.find_all("div", {"class":"a3b8729ab1 e6208ee469 cb2cbb3ccb"})
    if score:
        user_score = score[0].text.strip()
    else:
        user_score = 'NA'
    # Append user_rating to the user_ratings list
    review_score.append(user_score)
    
    ##################################### non-smoking #######################################################
    no_smoking = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    found_non_smoking_rooms = False

    for element in no_smoking:
        text = element.get_text()
        if "Non-smoking rooms" in text:
            found_non_smoking_rooms = True
            break  # Exit the loop early since we found "Non-smoking rooms" once

    if found_non_smoking_rooms:
        non_smoking = "yes"
    else:
        non_smoking = "no"

    Non_Smoking_Rooms.append(non_smoking)

    
    
    ################################### Star Rating ########################################
    quality = product_html.find_all("div", {"class": "a53cbfa6de c70c875b96"})
    quality_score = 'no'

    if quality:
        text = quality[0].text.strip()
        match = re.search(r'(\d out of 5)', text)
        if match:
            quality_score = match.group()
    star_rating.append(quality_score)
    
    
    #################################################### Family rooms #############################################
    fam = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    found_family_rooms = False

    for element in fam:
        text = element.get_text()
        if "Family rooms" in text:
            found_family_rooms = True
            break  # Exit the loop early since we found "Family rooms" once

    if found_family_rooms:
        fam_rooms = "yes"
    else:
        fam_rooms = "no"

    Family_rooms.append(fam_rooms)

    
    
    # ###################################### restorents ##############################################
    res = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    restaurant_found = False

    for element in res:
        text = element.get_text()
        if "restaurants" in text:
            restaurant_found = True
            break  # Exit the loop early since we found "restaurants" once

    if restaurant_found:
        restaurants = "yes"
    else:
        restaurants = "no"

    Restaurant.append(restaurants)

    
    ################################################ bar #################################################
    ba=product_html.find_all("div", {"class":"hotelchars"})[0].text
    # Use a regular expression to extract "Non-smoking rooms"
    matchsss = re.search(r'Bar', ba)

    if matchsss:
        bars = matchsss.group()
    else:
        bars = 'NA'
        
    bar.append(bars)
    
    ########################################## Heating ###################################
    he = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    heat_found = False

    for element in he:
        text = element.get_text()
        if "Heating" in text:
            heat_found = True
            break  # Exit the loop early since we found "Heating" once

    if heat_found:
        heating = "yes"
    else:
        heating = "no"

    Heating.append(heating)

    
    ###################################################### Lift ############################################
    li = product_html.find_all("div", {"class": "e2c5f6cdcb"})
    lift_found = False

    for element in li:
        text = element.get_text()
        if "Lift" in text:
            lift_found = True
            break  # Exit the loop early since we found "Lift" once

    if lift_found:
        lifts = "yes"
    else:
        lifts = "no"

    Lifts.append(lifts)
    
     ###################################### Single Bed ##################################################
    sb=product_html.find_all("div", {"class":"hotelchars"})[0].text
    # Use a regular expression to extract "Non-smoking rooms"
    matchsssss = re.search(r'single bed', sb)

    if matchsssss:
        single_bed = matchsssss.group()
    else:
        single_bed = 'NA'
        
    Single_Bed.append(single_bed)
    
    
    ############################################### double Bed ####################################################
    db=product_html.find_all("div", {"class":"hotelchars"})[0].text
    # Use a regular expression to extract "Non-smoking rooms"
    matchssssss = re.search(r'double bed', db)

    if matchssssss:
        double_bed = matchssssss.group()
    else:
        double_bed = 'NA'
        
    Double_bed.append(double_bed)
    
    ################################################### Break Fast ####################################
    
    # Find all elements that match the specified criteria
    elements = product_html.find_all("div", {"class": "bui-grid__column bui-grid__column-4 k2-hp--highlights"})

    # Check if any elements were found
    if elements:
        # If elements were found, access the first element and extract the text
        a = elements[0].text

        # Check if "Breakfast info" is present in the text
        if "Breakfast info" in a:
            # Split the text by the "Breakfast info" section and extract the breakfast information
            breakfast_section = a.split("Breakfast info")[1].strip()
            breakfast_info = breakfast_section.split("\n")[0].strip()
        else:
            breakfast_info = 'NA'
    else:
        # If no elements were found, set breakfast_info to 'NA'
        breakfast_info = 'NA'

    # Print or use the breakfast information as needed
    Breakfast.append(breakfast_info)

     
```

```python
import pandas as pd

# Define a list of DataFrame file names and corresponding city names
data = [
    (" Tokyo_dataset1.csv", "Tokyo"),
    ("Berlin_dataset1.csv", "Berlin"),
    ("Birmingham_dataset1.csv", "Birmingham"),
    ("DELHI_dataset1.csv", "New Delhi"),
    ("Dubai_dataset1.csv", "Dubai"),
    ("Kuala_Lampur_dataset1.csv", "Kuala Lumpur"),
    ("London_dataset1.csv", "London"),
    ("Manchester_dataset1.csv", "Manchester"),
    ("Melbourne_dataset1.csv", "Melbourne"),
    ("Paris_dataset1.csv", "Paris"),
    ("Sydney_dataset1.csv", "Sydney"),
    ("Toronto_dataset1.csv", "Toronto"),
]

# Create a list to hold the DataFrames
dataframes = []

# Loop through the data list, read each CSV file, and add the 'city' column
for file_name, city_name in data:
    df = pd.read_csv(file_name)
    df.insert(0, 'city', city_name)
    dataframes.append(df)

# Concatenate the DataFrames into a single DataFrame
new = pd.concat(dataframes, ignore_index=True)

# Set the index to start from 1
new.index = new.index + 1

# Display the combined DataFrame
new.head()
```

```python
new['number_of_reviews'] = new['number_of_reviews'].str.replace(',', '', regex=True).str.extract('(\d+)').fillna('0').astype(int)
new['Star_Rating'] = new['Star_Rating'].str.extract('(\d+)').astype(int)
new['bar'] = new['bar'].fillna('no').replace('Bar', 'yes')
# Split the 'Breakfast' column by commas into three separate Series
breakfast_split = new['Breakfast'].str.split(',', expand=True)
# Create new columns 'Breakfast Cuisine1,' 'Breakfast Cuisine2,' and 'Breakfast Cuisine3' in the DataFrame
new['Breakfast Cuisine1'] = breakfast_split[0].fillna('no')
new['Breakfast Cuisine2'] = breakfast_split[1].fillna('no')
new['Breakfast Cuisine3'] = breakfast_split[2].fillna('no')
new.drop("Breakfast",inplace=True,axis=1)
```
